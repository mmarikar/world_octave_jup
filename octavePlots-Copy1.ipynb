```octave
clear all
x = linspace(-4*pi, 4*pi, 100);
y1 = sin(x);
y2 = cos(x);
y3 = exp(x);
y4 = cosh(x);
subplot(3,3,1)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
subplot(3,3,2)
plot(x, y2, 'g')
xlabel('Radians')
ylabel('Unit Less')
title('cos')
subplot(3,3,3)
plot(x, y3, 'b')
xlabel('Radians')
ylabel('Unit Less')
title('exp')
axis([-3, 3, -1, 20])
subplot(3,3,4)
plot(x, y4, 'k')
xlabel('Radians')
ylabel('Unit Less')
title('cosh')
subplot(3,3,5)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
subplot(3,3,6)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
subplot(3,3,7)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
subplot(3,3,8)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
subplot(3,3,9)
plot(x, y1, 'r')
xlabel('Radians')
ylabel('Unit Less')
title('sin')
axis('equal')
axis([-3, 3, 0, 6])
```
